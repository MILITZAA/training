# 1 Add Data Sources

### CSV File
Download the CSV file and place it within our project directory.

1. Select **+ Create**.

2. Select **Data Source**.



3. Choose the data source origin. In this instance, we will select **File Upload**.



4. To finalize the process, locate our file and select **Open** to initiate the upload.

### From BigQuery

1. Repeat steps 1 and 2 as with CSV file ingestion.

2. Select the **BigQuery** option.



3. Select the **Custom Query** option. Subsequently, in the **Billing Project** field, enter `amco-data-analys`, and then input the SQL query to add the desired table.



4. Repeat step 3 for each additional table required.

# 2 Create Blended Data

To blend data:

1. Select **+ Create**.

2. Select **Report**.



3. Choose **BigQuery**.



4. Blend the tables that were previously added to Looker Studio from BigQuery. Then, select **CUSTOM QUERY**, enter the project name (in this case, `amco-data-analysis`), and finally, input the custom SQL query. In this instance, the query will blend the `bikeshare_stations` and `bikeshare_trips` tables.



5. Subsequently, Looker Studio will display a window where you can deploy a report using these blended tables.

### Looker Studio Anatomy

1. This is the report canvas where you build your visualizations.
2. The **DATA** panel displays the list of fields (variables). Here, you can also see each field's name and data type. From this panel, you can select and drag fields into **SETUP** panel (5) to configure your charts.
3. This section allows you to select a chart type from the available options.
4. These are examples of the chart types you can choose from.
5. In **SETUP** panel, you can configure field assignments such as Dimensions, Metrics, and Sort order. The **STYLE** panel allows you to customize the visual appearance of the chart, including the Chart title, Colors, Labels, Legend, Padding, Background, and Border. The available options within both panels will vary depending on the selected chart type.
6.  Within **SETUP** panel (5), the configuration options for the number of fields you can analyze (one, two, or three) will depend on the data types of the selected fields and the capabilities of the chosen chart type.



# 3 Create Calculated Fielda

1. In the bottom right corner, select the **Add a field** option.



2. Select the **Add calculated field"** option.



3. In the pop-up window, the variables are located on the left-hand side. On the right-hand side, where it says **Field name**, enter the desired name for the new variable. In the **Formula** field, enter the operation to be performed using the list of available variables.



Here is the result after adding two calculated fields. Those marked with the $fx$ symbol are the newly added fields.



# 4 Create Metrics

1. In the **Charts** option, select **Scorecard**.



2. In **Setup** and **Style** panels, customize the variable performance and appearance.





Here is the result of calculating and adding metrics such as:
* **Average Housing Price:** Computes the mean value of `SalePrice`.
* **Total Number of Homes Sold:** Counts the total number of entries in the dataset.
* **Average Living Area:** Calculates the average of `GrLivArea` (Gross Living Area above ground).
